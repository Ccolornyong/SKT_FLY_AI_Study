## Matplotlib

범주형 데이터 - 순위형 데이터는 좋고 나쁨이 있고 명목형은 없음

수치 데이터 - 연속형은 숫자 사이에 요소들이 존재할 수 있고 이산형은 없음

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# plot는 기본적으로 선 그래프를 그리는 함수
plt.plot(x, y, label = "Score")
plt.title("This is title")
plt.xlabel("X-value")
plt.ylabel("y-value")
plt.legend()
# 위의 오브젝트 리턴 값 안보임
plt.show()
```

```python
x = np.array([0,1,2,3,4])
y = np.array([0,2,4,6,8])
z = x**2

# plot는 기본적으로 선 그래프를 그리는 함수
plt.plot(x, y, label = "Score")
plt.plot(x, z,label="Score^2")
plt.title("This is title")
plt.xlabel("X-value")
plt.ylabel("y-value")
plt.legend()
# 위의 오브젝트 리턴 값 안보임
plt.show()
```

## Bar Chart

```python
x = ['A', 'B', 'C', 'D']
y = [10,30,20,40]

plt.bar(x, y)
```

```python
x = ['A', 'B', 'C', 'D']
y = [10,30,20,40]

plt.bar(x, y)

# 값만 오는게 아니라 위치도 같이 옴
for i,v in enumerate(y):
  plt.text(i, v+0.5, str(v), ha='center')

plt.ylim([0,50])
```

```python
plt.style.use('ggplot')

x = ['A', 'B', 'C', 'D']
y = [10,30,20,40]

plt.bar(x, y)

# 값만 오는게 아니라 위치도 같이 옴
for i,v in enumerate(y):
  plt.text(i, v+0.5, str(v), ha='center')

plt.ylim([0,50])
```

## 다중 플롯과 서브 플롯

```python
x = [0,1,2,3,4]
y = [1,4,5,9,16]
z = [0,-1,-4,-9,-16]
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

```python
# 새로 2개 가로 1개 그 중 첫번째를 그리겠다
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(x,z)
plt.show()
```

```python
# 하나 만들고 그림 그리는 방식
plt.subplot(2,2,1)
plt.plot(x,x)
plt.subplot(2,2,2)
plt.plot(x,z)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(y,y)
plt.show()
```

```python
# 한번에 만드는 방식
fig, axes = plt.subplots(2,2)
axes[0,0].plot(x,y, 'go--')
axes[1,1].plot(x,x)
plt.show()
```
