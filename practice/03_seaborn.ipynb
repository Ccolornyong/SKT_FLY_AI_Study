## Seaborn

lineplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_theme(style='darkgrid')
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
sns.lineplot(x=x, y=y)
sns.lineplot(x=x, y=z)
plt.title('sin and cos')
plt.show()
```

## Bar char

```python
cat = ['A', 'B', 'C', 'D']
val = [10,15,8,12]

sns.barplot(x=cat, y=val, palette='viridis')
for i, v in enumerate(val):
  plt.text(i, v-5, str(v), fontsize = 15, ha = 'center', color = 'white') # x, y, 

plt.show()
```

```python
cat = ['A', 'B', 'C', 'D']
val = np.array([10,15,8,12])
val = np.sort(val)

sns.barplot(x = val, y = cat)
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)
sns.histplot(x=data, bins = 40, kde=True)
plt.show()
```

## Scatterplot

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.scatterplot(x=df['petal_length'], y=df['petal_width'], hue = df['species'])
plt.show()
```

```python
df.columns
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width',
                data = df,
                hue = 'species',
                style='species')
plt.show()
```

```python
sns.pairplot(data=df, hue='species')
plt.show()
```

## Box plot

```python
sns.boxplot(data=df, x='species', y='sepal_width', hue='species')
plt.show()
```

## violinplot

```python
sns.violinplot(data=df, x='species', y='sepal_width', hue='species')
plt.show()
```
