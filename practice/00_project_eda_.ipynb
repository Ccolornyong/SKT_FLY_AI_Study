# 데이터 불러오기

```python
df = pd.read_csv("/content/drive/MyDrive/SKT_FLY_AI/Dataset/국가보훈부_독립유공자 명단_20241219.csv", encoding = 'cp949')
df
```

```python
df
```

```python
df['사망년월일'].unique()
```

```python
df['생년월일'].unique()
```

```python
# 사망년월일 미상인 사람 삭제
idx = df[df['사망년월일'] == '미상'].index
df.drop(idx, inplace=True)
```

```python
# 사망년월일 미상인 사람 삭제
idx = df[df['사망년월일'] == '생존'].index
df.drop(idx, inplace=True)
```

```python
# 생년월일 미상인 사람 삭제
idx = df[df['생년월일'] == '미상'].index
df.drop(idx, inplace=True)
```

```python
# 생년월일 미상인 사람 삭제
idx = df[df['생년월일'] == '미공개'].index
df.drop(idx, inplace=True)
```

```python
df['생년월일'].str.split('-').str[0]
```

```python
# 생년월일 및 사망년도 년도만 남기기
df['생년월일'] = df['생년월일'].str.split('-').str[0]
df['사망년월일'] = df['사망년월일'].str.split('-').str[0]
```

```python
df
```

```python
df.dtypes
```

```python
df[['생년월일', '사망년월일']] = df[['생년월일', '사망년월일']].apply(pd.to_numeric)
```

```python
df['death'] = df['사망년월일'] - df['생년월일']
df
```

## 운동계열별 평균 수명

```python
df1 = df.groupby('운동계열')['death'].mean()
```

```python
df1
```

```python
sns.barplot(data=df1)
plt.xticks(rotation = 45)
plt.show()
```

## 사망 나이에 따른 훈장 가치 비교

```python
df['death']%10
```

```python
df['age_group'] = df['death']%10
```

```python
df
```

```python
df.groupby('age_group')['훈격(소분류)'].value_counts()
```

```python
sns.countplot(x = df['훈격(소분류)'], hue=df['age_group'])
plt.show()
```
