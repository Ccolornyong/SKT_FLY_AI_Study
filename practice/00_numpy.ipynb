##파이썬 리스트와 넘파이 속도 비교

```python
import numpy as np
```

```python
python_list = [i for i in range(1000000)]
```

```python
arr = np.array(python_list)
```

```python
# 매직 커멘드
%%timeit
sum(python_list)
```

```python
%%timeit
np.sum(arr)
```

## 인덱싱(indexing)과 슬라이싱(slicing)

```python
arr[3]
```

```python
arr[-1]
```

# 불리언 인덱싱(Boolean indexing)

```python
arr = np.array([1,2,3,4,5])
b = [False, True, False, True, False]
arr[1:3]
# 가지고 오고 싶은 곳에는 True 아닌 곳에는 False
arr[b]
```

```python
# 조건에 맞춰서 요소 가져오기
arr > 3
```

```python
arr[arr>3]
```

```python
arr2 = np.array([[1,2,3],
                [4,5,6]])
```

```python
# 일차원으로 바꿤요
arr2[arr2>2]
```

```python
arr = np.arange(10)
arr
```

```python
# 얘는 끝까지 전부다 프린트
arr[3:]

# 얘는 마지막 앞까지 읽어옴
arr[3:-1]

# 0부터 끝까지
arr[:]

# 6부터 끝까지
arr[-4:]
```

## 배열의 연산

```python
a = [1,2,3]
b = [3,4,5]
```

```python
# 파이썬 리스트는 두개의 리스트를 단순 결합하기만 함
a + b
```

```python
a = np.array([1,2,3])
b = np.array([3,4,5])
```

```python
# 넘파이의 어레이는 같은 위치에 있는 요소를 더 함
a + b
```

```python
# 브로드캐스팅
# 1이 내부적으로 [1,1,1]로 바뀜
a + 1
```

```python
arr = np.array([1,2,3,4])
arr
```

```python
arr.sum(), arr.mean()
```

```python
np.sum(arr), np.mean(arr)
```

```python
arr2 = np.array([[1,3,2],
                 [2,3,4]])
arr2
```

```python
arr2.sum()
arr2.sum(axis=0)
arr2.sum(axis=1)
```

```python
arr2.mean()
arr2.mean(axis=0)
arr2.mean(axis=1)
```

```python
arr2 = np.array([[1,3,2],
                 [2,3,4]])
arr2
```

```python
arr2.shape
```

```python
# shape을 다시 바꾸고 싶을 때
arr2.reshape(3,2)
arr2.T
```

```python
# 삼차원으로 바꾸기
# 하나 짜리가 2개 모여 있는게 3개 있음
arr2.reshape(3,2,1)
```

```python
# 위는 2차원으로 새로 모양 한 줄, 아래는 1차원으로 가로 모양 한 줄
arr2 = [[1],[2],[3]]
arr2 = [1,2,3]
```
